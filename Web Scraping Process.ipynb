## Web Scraping Process

## Steps
1. Identify the url
2. Extract HTML code from Url
3. Parsed HTML code to BeautifulSoup object.
4. Inspect the web page and find the HTML tags for data that you want to Extract.
5. Using Beautful Soup object, extract the tags you are interested in.            

## importing necessory libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

### Fetching url

```python
url = "https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=13&Tx_State=MP&Tx_District=57&Tx_Market=8205&DateFrom=01-Apr-2025&DateTo=04-Aug-2025&Fr_Date=01-Apr-2025&To_Date=04-Aug-2025&Tx_Trend=2&Tx_CommodityHead=Soyabean&Tx_StateHead=Madhya+Pradesh&Tx_DistrictHead=Ashoknagar&Tx_MarketHead=Chanderi"
response = requests.get(url)
```

### Parse HTML

```python
soup = BeautifulSoup(response.text, "html.parser")
```

### Find the div containing the data 

```python
table_div = soup.find("div", id="cphBody_UpdatePanel1")
rows = []
if table_div:
    # Find all rows inside the div
    for row in table_div.find_all("tr"):
        cols = [td.text.strip() for td in row.find_all(["td", "th"])]
        if cols:   # avoid empty rows
            rows.append(cols)
else:
    print("Error: Table not found")
```

### Convert data into DataFrame 

```python
df = pd.DataFrame(rows[1:], columns=rows[0])
```
